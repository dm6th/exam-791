```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Assuming 'df' is your DataFrame and 'PropertyPrice' is the column for property prices

# Extract property prices
property_prices = df['MEDV']

# Plot histogram with 40 bins
plt.figure(figsize=(10, 6))
count, bins, ignored = plt.hist(property_prices, bins=40, density=True, alpha=0.6, color='skyblue', edgecolor='black')

# Calculate mean and standard deviation
mean_price = property_prices.mean()
std_price = property_prices.std()

# Generate normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 1000)
p = norm.pdf(x, mean_price, std_price)

# Plot the normal distribution curve
plt.plot(x, p, 'r--', linewidth=2, label='Normal Distribution')

# Add labels and title
plt.xlabel('Property Price')
plt.ylabel('Density')
plt.title('Property Price Density with Normal Distribution')
plt.legend()
plt.grid()

# Show plot
plt.show()
```

```python
#Data Transformations: Log Transformation [6 Marks]
#Plot the property price with the log transformation normality distribution curve using a histogram with
#a title: “Property price with log transformation”. (3 marks)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Assuming 'PropertyPrice' is the column in your DataFrame 'df'
# Apply log transformation
log_property_prices = np.log1p(df['MEDV'])  # log1p handles log(0) safely

# Plot histogram with 40 bins
plt.figure(figsize=(10, 6))
count, bins, ignored = plt.hist(log_property_prices, bins=40, density=True, alpha=0.6, color='skyblue', edgecolor='black')

# Calculate mean and standard deviation of the log-transformed data
mean_log_price = log_property_prices.mean()
std_log_price = log_property_prices.std()

# Generate normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 1000)
p = norm.pdf(x, mean_log_price, std_log_price)

# Plot the normal distribution curve
plt.plot(x, p, 'r--', linewidth=2, label='Normal Distribution')

# Add labels, title, and legend
plt.xlabel('Log of Property Price')
plt.ylabel('Density')
plt.title('Property Price with Log Transformation')
plt.legend()
plt.grid()

# Show plot
plt.show()
```
